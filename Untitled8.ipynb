```python
#Direktori Isi Latih/Training

train_girl = os.path.join(train_dir, 'girl/')
train_boy = os.path.join(train_dir, 'boy/')

#Direktori Isi Validasi

validation_girl = os.path.join(validation_dir, 'girl/')
validation_boy = os.path.join(validation_dir, 'boy/')
```

```python
import random
from shutil import copyfile

def train_val_split(source, train, val, train_ratio):
    total_size = len(os.listdir(source))
    train_size = int(total_size * train_ratio)
    val_size = total_size - train_size

    randomized = random.sample(os.listdir(source), total_size)
    train_files = randomized[0:train_size]
    val_files = randomized[train_size:total_size]

    for i in train_files:
        i_file = os.path.join(source, i)  # Gunakan os.path.join
        destination = os.path.join(train, i)  # Gunakan os.path.join
        copyfile(i_file, destination)

    for i in val_files:
        i_file = os.path.join(source, i)
        destination = os.path.join(val, i)
        copyfile(i_file, destination)

#jumlah pembagian data training dan validasi
train_ratio = 0.9 # This line was incorrectly indented

#pembagian training dan validasi
# Training
source_00 = girl_dir
train_00 = train_girl
val_00 = validation_girl
train_val_split(source_00, train_00, val_00, train_ratio)

# Validation
source_01 = boy_dir
train_01 = train_boy
val_01 = validation_boy
train_val_split(source_01, train_01, val_01, train_ratio)
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.3,
                                   zoom_range = 0.1,
                                   horizontal_flip = True,
                                   rotation_range = 30,
                                   fill_mode ='nearest',
                                   height_shift_range = 0.2, # Corrected argument name
                                   width_shift_range = 0.2,) # Corrected argument name

val_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.3,
                                   zoom_range = 0.1,
                                   horizontal_flip = True,
                                   rotation_range = 30,
                                   fill_mode ='nearest',
                                   height_shift_range = 0.2, # Corrected argument name
                                   width_shift_range = 0.2,) # Corrected argument name
```

```python
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size = (150, 150),
                                                    batch_size = 10,
                                                    class_mode = 'categorical')

val_generator = val_datagen.flow_from_directory(validation_dir,
                                                target_size = (150, 150),
                                                batch_size = 10,
                                                class_mode = 'categorical')
```

```python
class myCallBack(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.99):
      print("\nAkurasi telah mencapai >99%!")
      self.model.stop_training = True
callbacks = myCallBack()
```

```python
model = tf.keras.models.Sequential([
tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(150, 150, 3)),
tf.keras.layers.MaxPooling2D(2, 2),
tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
tf.keras.layers.MaxPooling2D(2, 2),
tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
tf.keras.layers.MaxPooling2D(2, 2),
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(200, activation='relu'),
tf.keras.layers.Dropout(0.3, seed=112),
tf.keras.layers.Dense(500, activation='relu'),
tf.keras.layers.Dropout(0.5, seed=112),
tf.keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.summary()
```

```python
history = model.fit(train_generator,
                    epochs = 1000,
                    steps_per_epoch = 6,
                    validation_data = val_generator,
                    validation_steps = 1,
                    verbose = 1,
                    callbacks = [callbacks]
)
```

```python
print("validation_dir:", validation_dir)
print("validation_girl:", validation_girl)
```

```python
model.compile(loss = "categorical_crossentropy",
              optimizer = 'Adam',
              metrics = ['accuracy'])
```

```python
%matplotlib inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label = 'Training Accuracy')
plt.plot(epochs, val_acc, 'b', label = 'Validation Accuracy')
plt.title('Training and Validation accuracy')
plt.legend(loc = 'best')
plt.show()

plt.plot(epochs, loss, 'r', label = 'Training Loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation Loss')
plt.title('Training and Validation accuracy')
plt.legend(loc = 'best')
plt.show()
```
